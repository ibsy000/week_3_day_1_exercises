### Exercise 2 - Write a Python class for an Animal that has a name and energy attributes. The animal class should also have methods for eat, sleep, and play that will take in an integer and increase/decrease the energy of the animal with a formatted print statement

```python
# Example 1
# buddy = Animal('Buddy', 10)
# buddy.play(5) -> "Buddy is playing for 5 minutes. His energy is now 5"
# buddy.sleep(10) -> "Buddy is sleeping for 10 minutes. His energy is now 15"
```

```python
class Pokemon:
    
    def __init__(self, name, energy = 15):
        self.name = name
        self.energy = energy
        
    
    def eat(self, candy_amt):
        self.energy += candy_amt
        print(f"Your {self.name} just consumed {candy_amt} candy! It's energy is now {self.energy}. That's sugar!")
    
    
    def play(self, hours_of_play):
        self.energy -= hours_of_play
        print(f"It's time to play! For {hours_of_play} hours of play, your {self.name} will now have {self.energy} energy points.")
    
    
    def sleep(self, hours_of_sleep):
        self.energy += hours_of_sleep
        print(f"Yawwwwn...{self.name} is tried! But after sleeping for {hours_of_sleep} hours, {self.name}'s energy is now up to {self.energy}!")


pokemon1 = Pokemon('Psyduck', 5) # animal that has name and energy attribute
# pokemon2 = Pokemon('Cubone')

pokemon1.eat(5)
# pokemon2.eat(3)

pokemon1.play(2)
# pokemon2.play(8)

pokemon1.sleep(10)
# pokemon2.sleep(6)
```
