### Exercise 1 - Turn the shopping cart program into an object-oriented program

```python
from IPython.display import clear_output
import time

# Create a class called cart that retains items and has methods to add, remove,
# and show


# Item class to represent the first instance of adding an item to the shopping cart
class Item:

    def __init__(self, user_name):

        self.user_name = user_name
        self.shopping_cart = {}
        
    
    def add_to_cart(self):
        
        clear_output()
        
        item_name = input(f"Please add an item to your shopping cart, {self.user_name}: ")
        item_name = item_name.strip()
        self.shopping_cart[item_name] = {}

        item_price = input(f"What is the price for {item_name}? ")
        item_price = float(item_price)
        self.shopping_cart[item_name]['item_price'] = item_price

        item_quantity = input(f"How many {item_name}(s) would you like? ")
        item_quantity = int(item_quantity)
        self.shopping_cart[item_name]['item_quantity'] = item_quantity
        
        clear_output()

        print(f"{item_quantity} {item_name}(s) have been added to your cart.")


    def show_cart(self):

        clear_output()
        
        grand_total = []

        if self.shopping_cart:

            print("Your current cart:")

            for item in self.shopping_cart.keys():

                print(f"\t{self.shopping_cart[item]['item_quantity']} {item}(s) x {self.shopping_cart[item]['item_price']:.2f}")
                total_price = self.shopping_cart[item]['item_price'] * self.shopping_cart[item]['item_quantity']
                grand_total.append(total_price)

            print(f"Total price: {sum(grand_total):.2f}")

        else:
            print("There are no items in your cart at this time.")


    def remove_item(self):
        
        clear_output()
        
        remove = input("Which item would you like to remove? ")

        if remove not in self.shopping_cart.keys():

            print(f"{remove} is not in your cart.")

        elif remove in self.shopping_cart.keys():
            
            remove_quantity = input(f"How many of the {self.shopping_cart[remove]['item_quantity']} {remove}(s) would you like to remove? ")
            remove_quantity = int(remove_quantity)

            new_quantity = self.shopping_cart[remove]['item_quantity'] - remove_quantity
            self.shopping_cart[remove]['item_quantity'] = new_quantity

            print(f"{remove_quantity} {remove}(s) successfully removed from your cart.")
        

    def stick_up(self):
        
        clear_output()

        print(f'{self.user_name}: "This is a stickup!! Put your hands in the air!"')
        time.sleep(4)

        print(f'Apu: "No, please don\'t do this! I have a wife and eight children!"')
        time.sleep(4)

        print(f'{self.user_name}: "We can do this the easy way or the hard way!"')
        time.sleep(4)

        print(f'Apu: "I choose the hard way, {self.user_name}!"')
        time.sleep(2)

        print(f"*Apu quickly reaches under the counter, grabs a fully loaded shotgun and points it at {self.user_name}*")
        time.sleep(3)

        print(f'Apu: "You wanna dance, punk?"')
        time.sleep(4)

        print(f'{self.user_name}: "Uhh actually, you know what? I totally forgot I have to be somewhere.."')
        time.sleep(3)

        print(f'*{self.user_name} runs out of the Kwik-E-Mart with their tail tucked between their legs*')
        time.sleep(4)


def main():

    print("Welcome to the Kwik-E-Mart. My name is Apu, I am the owner of this shop.")

    user_name = input("What is your name? ")
    user_name = user_name.title().strip()
    
    clear_output()

    user_cart = Item(user_name)

    print(f"Here is a shopping cart to get you started, {user_name}.")

    while True:

        next = input("What would you like to do? add/remove/show/quit/stickup ")

        if next == 'quit':
            print("Thank you! Come again!")
            break  

        elif next == 'add':
            user_cart.add_to_cart()

        elif next == 'remove':
            user_cart.remove_item()

        elif next == 'show':
            user_cart.show_cart()

        elif next == 'stickup':
            user_cart.stick_up()

        else:
            print("That is not a valid response.")

main()
```

### Exercise 2 - Write a Python class for an Animal that has a name and energy attributes. The animal class should also have methods for eat, sleep, and play that will take in an integer and increase/decrease the energy of the animal with a formatted print statement

```python
# Example 1
# buddy = Animal('Buddy', 10)
# buddy.play(5) -> "Buddy is playing for 5 minutes. His energy is now 5"
# buddy.sleep(10) -> "Buddy is sleeping for 10 minutes. His energy is now 15"
```

```python
class Pokemon:
    
    def __init__(self, name, energy = 15):
        self.name = name
        self.energy = energy
        
    
    def eat(self, candy_amt):
        self.energy += candy_amt
        print(f"Your {self.name} just consumed {candy_amt} candy! It's energy is now {self.energy}. That's sugar!")
    
    
    def play(self, hours_of_play):
        self.energy -= hours_of_play
        print(f"It's time to play! For {hours_of_play} hours of play, your {self.name} will now have {self.energy} energy points.")
    
    
    def sleep(self, hours_of_sleep):
        self.energy += hours_of_sleep
        print(f"Yawwwwn...{self.name} is tried! But after sleeping for {hours_of_sleep} hours, {self.name}'s energy is now up to {self.energy}!")


pokemon1 = Pokemon('Psyduck', 5) # animal that has name and energy attribute
# pokemon2 = Pokemon('Cubone')

pokemon1.eat(5)
# pokemon2.eat(3)

pokemon1.play(2)
# pokemon2.play(8)

pokemon1.sleep(10)
# pokemon2.sleep(6)
```
